```python
# Import modules
import os 
import csv
import pandas as pd 

# Point to file 

spy_data="./SPY.csv"
dia_data="./DIA.csv"

# Create dataframes for data

spy_df=pd.read_csv(spy_data)
dia_df=pd.read_csv(dia_data)

# Pull Columns

spy_close=spy_df["Close"]
dia_close=dia_df["Close"]
spy_close_df=pd.DataFrame(spy_close, columns=["Close"])
dia_close_df=pd.DataFrame(dia_close, columns=["Close"])

# Rename Columns for merging

renamed_spy_close_df=spy_close_df.rename(columns={"Close":"SPY Close"})
renamed_dia_close_df=dia_close_df.rename(columns={"Close":"DIA Close"})

rounded_renamed_dia_close_df=round(renamed_dia_close_df, 2)
rounded_renamed_spy_close_df=round(renamed_spy_close_df, 2)
# Combine Columns

compare_close_df=rounded_renamed_dia_close_df.join(rounded_renamed_spy_close_df, lsuffix="_left", rsuffix="_right")

# Percent Change
dia_perc_change_df=renamed_dia_close_df.pct_change()
dia_final_change_df=round(dia_perc_change_df*100, 2)

spy_perc_change_df=renamed_spy_close_df.pct_change()
spy_final_change_df=round(spy_perc_change_df*100, 2)

renamed_spy_final_change_df=spy_final_change_df.rename(columns={"SPY Close":"SPY Weekly Change(%)"})
renamed_dia_final_change_df=dia_final_change_df.rename(columns={"DIA Close":"DIA Weekly Change(%)"})

compare_perc_change_df=renamed_dia_final_change_df.join(renamed_spy_final_change_df, lsuffix="_left", rsuffix="_right")

# Master chart
final_chart_df=compare_close_df.join(compare_perc_change_df, lsuffix="_left", rsuffix="_right")
final_chart_df
```
