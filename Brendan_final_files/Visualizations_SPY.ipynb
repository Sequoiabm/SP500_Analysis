```python
#Import the necessary dependencies/libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#Read the raw data from the SPY and DIA csv files
spy_csv_df= "./SPY.csv"
spy_df = pd.read_csv(spy_csv_df)
spy_csv_past_year_df = "./SPY_past_year.csv"
spy_past_year_df = pd.read_csv(spy_csv_past_year_df)
spy_csv_weekly_df = "./SPY_weekly.csv"
spy_weekly_df = pd.read_csv(spy_csv_weekly_df)
dia_csv_df= "./DIA.csv"
dia_df = pd.read_csv(dia_csv_df)
dia_csv_past_year_df = "./DIA_past_year.csv"
dia_past_year_df = pd.read_csv(dia_csv_past_year_df)

#Create line graph for SPY stock close price from 2001-2023 for each month
months = np.arange(0, 264, 1)
close_price = spy_df["Close"]
plt.figure(figsize=(10, 10))
plt.plot(months, close_price, color="blue", linewidth=1, label="Close")
plt.ylim(0, 500)
plt.title("SPY price trend over time (2001-2023)")
plt.xlabel("Months")
plt.ylabel("Price")
plt.tight_layout()
plt.grid()
plt.savefig("./visual_spy_price.png")
plt.show()
```

```python
#Create weekly change line graph for SPY in the past year
weeks = np.arange(0, 52, 1)
weekly_change= spy_weekly_df["SPY Weekly Change(%)"]
plt.figure(figsize=(10, 10))
plt.plot(weeks, weekly_change, color="purple", linewidth=1, label="Weekly Change")
plt.ylim(-10, 10)
plt.title("SPY weekly change (2022-2023)")
plt.xlabel("Weeks")
plt.ylabel("Weekly Change (%)")
plt.tight_layout()
plt.grid()
plt.savefig("./visual_spy_change.png")
plt.show()
```

```python
#Create volume trend line graph for SPY in the past year
months = np.arange(0, 12, 1)
volume = spy_past_year_df["Volume"]
plt.figure(figsize=(10, 10))
plt.plot(months, volume, color="green", marker = "s", linewidth=1)
plt.ylim(0, 2800000000)
plt.title("SPY volume trend over time (2022-2023)")
plt.xlabel("Months")
plt.ylabel("Volume")
plt.tight_layout()
plt.grid()
plt.savefig("./visual_spy_volume.png")
plt.show()
```

```python
#Create line graph for DIA close price from 2001-2023 for each month
months = np.arange(0, 264, 1)
close_price = dia_df["Close"]
plt.figure(figsize=(10, 10))
plt.plot(months, close_price, color="blue", linewidth=1, label="Close")
plt.ylim(0, 500)
plt.title("DIA price trend over time (2001-2023)")
plt.xlabel("Months")
plt.ylabel("Price")
plt.tight_layout()
plt.grid()
plt.savefig("./visual_dia_price.png")
plt.show()
```
